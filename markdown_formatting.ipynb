# Grundlegende Formatierungen
## Textstil
- Fett: **Text**
- Kursiv: *Text*
- Fett + kursiv: ***Text***
- Durchgestrichen: ~~Text~~
- Inline-Code: `code`
- Blockzitat: > Text


## Überschriften
# H1
## H2
### H3
#### H4
##### H5
###### H6



## Listen
Unsortiert
- Punkt
* Punkt
+ Punkt


### Sortiert
1. Erster
2. Zweiter


### Verschachtelt
- Punkt
  - Unterpunkt
    - Tiefer


## Links & Bilder
### Link
- [Text](https://example.com)

### Bild
- ![Alt-Text](https://example.com/bild.png)
